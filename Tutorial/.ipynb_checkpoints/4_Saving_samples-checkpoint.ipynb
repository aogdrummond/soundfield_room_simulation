## Saving simulated room as .mat file

The soundfield simulated for the room is saved on a .mat format, aiming to allow it to be used as sample for neural network.

- To save the sample just set "save" parameter as True, inserting on dataset_path a string defining the path of where the data will be saved

- The file will be named according to the following standard:\

numberOfSamplesOnDirectory_d_RoomDim0_RoomDim1_RoomDim2_s_SourceCoordx_SourceCoordy_.mat\

Ex: 0_d_1.3_2.5_6.5_s_0.0_2.0_.mat

```python
sample_path = "\\Tutorial\\sample\\0_d_1.3_2.5_6.5_s_0.0_2.0_.mat"
mat = scipy.io.loadmat("".join([str(Path(os.getcwd()).parent),sample_path]))
mat.keys()
```

Each file contains not only the sound pressure on each receiver (AbsFrequencyResponse) but also some metadata required for it to be calculated like "Frequency" and "Receiver_Coord", as well as attributes not required but that were included to the file be consistent with the samples of the original paper about neural networks, like "Mu" and "Psi_r", which used analytic methods to calculate it.

The structure for FrequencyResponse/AbsFrequencyResponse depends on what was set as parameter for the sample. On default it is set 32 rows of receivers (x_receiver attribute) and 32 receivers on each row (y_receiver attribute), as well as the 151 frequencies from 0 to 150 whose pressure will be stored. Therefore the data will be an array (32x32x151), which fits to neural networks requirement. This shape may be changed deepending on the needs.
